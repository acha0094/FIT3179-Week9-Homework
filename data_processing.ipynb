```python
import pandas as pd
```

```python
deaths = pd.read_csv('data/epidemic/deaths_state.csv')
```

```python
func = {'deaths_new': 'sum'}
```

```python
deaths_df = deaths.groupby(['state']).agg(func)
deaths_df.rename(columns = {'deaths_new': 'total_deaths'}, inplace=True)
# groupbyCountry.sort_values(by='total_deaths', ascending=False, inplace=True)
deaths_df = deaths_df.rename(index={'W.P. Kuala Lumpur': 'Kuala Lumpur', 'W.P. Labuan': 'Labuan', 'W.P. Putrajaya': 'Putrajaya'})
deaths_df
```

```python
population_df = pd.read_csv('data/static/population.csv')
population_df
```

```python
population_df.drop(columns=['idxs', 'pop_18', 'pop_60', 'pop_12', 'pop_5'], inplace=True)
population_df.rename(columns = {'pop': 'population'}, inplace=True)
population_df
```

```python
population_df['state'] = population_df['state'].replace(['W.P. Kuala Lumpur', 'W.P. Labuan', 'W.P. Putrajaya'], ['Kuala Lumpur', 'Labuan', 'Putrajaya'])
population_df
```

```python
df = pd.merge(deaths_df, population_df, on='state')
df
```

```python
df.to_csv('data/epidemic/total_deaths_state.csv', index=False)
```
